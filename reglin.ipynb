```python
import os 
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np 
import argparse
```

```python
mergeddata_CN = mergeddata[mergeddata['Group'] == 'CN']
mergeddata_MCI = mergeddata[mergeddata['Group'] == 'MCI']
mergeddata_AD = mergeddata[mergeddata['Group'] == 'AD']
```

```python
fig = plt.figure(figsize = (10, 7))
sns.regplot(x=mergeddata_MCI.NPICSEV, y = mergeddata_MCI.temporal_parahippocampal_ThickL_DK, color = 'blue')
sns.regplot(x=mergeddata_AD.NPICSEV, y = mergeddata_AD.temporal_parahippocampal_ThickL_DK, color = 'orange')

plt.legend(labels=['MCI', 'AD'])
plt.title('Relationship between NPS severity and brain morphology', size=24)
plt.xlabel('NPICSEV (Agitation)', size=18)
plt.ylabel('temporal_parahippocampal_Thick_lh_DK', size=18);
plt.axis([0,3, 0, 4])
```
